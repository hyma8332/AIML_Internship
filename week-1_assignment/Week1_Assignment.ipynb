## Import Libraries

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("C:/Users/DELL/Downloads/student_marks (1).csv")
df.head()
```

## Calculate Average Marks

```python
subjects = ["Maths","Physics","Chemistry","English","Biology","Economics","History","Civics"]
df["Average"] = df[subjects].mean(axis=1)
df.head()
```

## Descriptive Statistics

```python
mean = df["Average"].mean()
median = df["Average"].median()
mode = df["Average"].mode()[0]
std = df["Average"].std()
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std)
```

## Complete Statistics

```python
print(df["Average"].describe())
```

## Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(df["Average"], bins=5, edgecolor='black')
plt.title("Distribution of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Number of Students")
plt.show()
```

## Boxplot

```python
plt.figure(figsize=(5,6))
plt.boxplot(df["Average"])
plt.title("Box Plot of Average Marks")
plt.ylabel("Average Marks")
plt.show()
```

## Interpretation
The dataset contains marks of 10 students in eight different subjects. The average marks of each student were calculated and analyzed. The mean and median values are close, indicating that the students' performance is fairly balanced. The standard deviation shows a moderate variation in average marks among students. The histogram shows that most students scored around the middle range of average marks. The box plot indicates whether any outliers are present. Overall, the class performance is consistent with no extreme differences in average scores.
